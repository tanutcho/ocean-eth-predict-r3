## Let's started

We use testing set as last 7 days for 12 hours prediction

# Fear and Greed Index Data Acquisition


```python
#Fear and Greed Index Acquire from Alternative.me with 1500 data points
url = "https://api.alternative.me/fng/?limit=1500"

# Send the API request and get the response
response = requests.get(url)
# Extract the Fear and Greed Index data from the response
dat = response.json()['data']

fng_df = pd.DataFrame(columns=['date', 'fng_value'])

# Add value for each day to dataframe
for index in dat:
    index_date = datetime.fromtimestamp(int(index['timestamp'])).strftime('%Y-%m-%d %H:%M:%S')
    index_value = index['value']
    fng_df = fng_df.append({'date': index_date, 'fng_value': index_value}, ignore_index=True)
```

```python
# Convert to numeric

fng_df['fng_value'] = pd.to_numeric(fng_df['fng_value'])
```

```python
#Recheck that the dataframe match what we need and selected since 2020-01-01 
fng_df['date'] = pd.to_datetime(fng_df['date'],format='%Y-%m-%d %H:%M:%S')
fng_df = fng_df[fng_df['date']>='2020-01-01']

#Sort it well
fng_df = fng_df.sort_values(by='date')
fng_df.info()
```

# ETH/USDT Price Data Acquisition

```python
ex=ccxt.binance()
from_ts = ex.parse8601('2020-01-01 00:00:00')
ohlcv_list = []
ohlcv = ex.fetch_ohlcv('ETH/USDT', '1h', since=from_ts, limit=1000)
ohlcv_list.append(ohlcv)
while True:
    from_ts = ohlcv[-1][0]
    new_ohlcv = ex.fetch_ohlcv('ETH/USDT', '1h', since=from_ts, limit=1000)
    ohlcv.extend(new_ohlcv)
    if len(new_ohlcv)!=1000:
    	break
```

```python
_data = pd.DataFrame(ohlcv, columns=['date', 'open', 'max', 'min', 'close', 'volume'])
_data['date'] = pd.to_datetime(_data['date'],unit='ms')
```

```python
_data
```

# Merge data altogether

```python
#Merge data altogether for match nearest points 
data = pd.merge_asof(_data,fng_df, direction='nearest', left_on='date', right_on='date')
```

# Train,Test,Validation split set 

```python
data = data.drop_duplicates(subset=['date'])

train_set = data.loc[data['date'] <= '2023-01-31'].drop_duplicates(subset=['date'])
val_set = data.loc[(data['date'] >= '2023-02-01') & (data['date'] <= '2023-02-17')].drop_duplicates(subset=['date'])
```

```python
test_set = data.loc[(data['date'] > '2023-02-10')].drop_duplicates(subset=['date'])
```

```python
train_set = train_set[['date','close','fng_value']]
val_set = val_set[['date','close','fng_value']]
train_set.columns = ['ds','y','fng']
val_set.columns = ['ds','y','fng']
```

```python
test_set = test_set[['date','close','fng_value']]
test_set.columns = ['ds','y','fng']
```

```python
print(train_set.shape)
print(test_set.shape)
print(val_set.shape)
```

## Let's training and forecasting with NeuralProphet
To using this we training with 7 days lagged data (Auto-regression) and trying to forecast next 24 hours
with 2 hidden layers of neural network, trend and seasonality

We use fear and greed index as the lagged regressor features for prediction

```python
from neuralprophet import NeuralProphet

trainer_config = {"accelerator":"cuda"}

```

```python

m = NeuralProphet(n_forecasts=24,
        n_lags=24*7,
        num_hidden_layers =1,
        seasonality_mode='multiplicative',
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=True,
                  n_changepoints=20,
                  trend_reg=1,
                 epochs=100,
                  batch_size=256
                 )
m.add_lagged_regressor('fng')

```

```python
metrics = m.fit(train_set, freq="h",validation_df=val_set)
```

```python
plt.plot(metrics['RMSE'],label="Train RMSE")
plt.plot(metrics['RMSE_val'],label="Val RMSE")
plt.legend()
```
